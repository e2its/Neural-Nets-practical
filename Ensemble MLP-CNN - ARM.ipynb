# Dataset ARM: Ensemble MLP-CNN para datos basados en Time Series

```python
import pandas

## Checking tensorflow backend engine
import tensorflow as tf
message = tf.constant('Hello world!')
session = tf.Session()
session.run(message)
import keras.backend as K
print(K.epsilon())

from keras.models import Model
from keras.layers import Dense, Activation, Dropout, Input, Conv2D, MaxPooling2D, Flatten
from keras.layers.merge import concatenate
from keras.utils import to_categorical, plot_model

from numpy import argmax,reshape
```

```python
train_df = pandas.read_csv("https://raw.githubusercontent.com/e2its/datasets/master/ARM-Metric-train-TS.csv",encoding="UTF-8")
```

```python
test_df = pandas.read_csv("https://raw.githubusercontent.com/e2its/datasets/master/ARM-Metric-test-TS.csv",encoding="UTF-8")
```

### Normalizamos los datos a mean= 0 y std = 1

```python
from sklearn.preprocessing import StandardScaler as sc
```

```python
encoder = dict()
decoder = dict()
counter = 0

for itera in train_df['ATYPE'].unique():
    encoder[itera] = counter
    decoder[counter] = itera
    counter += 1
```

```python
train_df['ATYPE'] = train_df['ATYPE'].apply(lambda x: encoder[x])
test_df['ATYPE'] = test_df['ATYPE'].apply(lambda x: encoder[x])
```

```python
train_df['ATYPE']
```

```python
columns = list(train_df.columns)
```

```python
columns.remove('ATYPE')
```

```python
scaler = sc(copy=False)
scaler.fit(train_df[columns])
```

```python
train_df[columns] = scaler.transform(train_df[columns])
```

```python
test_df[columns] = scaler.transform(test_df[columns])
```

```python
train_df.groupby('ATYPE').mean()
```

```python
test_df.groupby('ATYPE').mean()
```

```python
x_train = train_df[columns].to_numpy()
y_train = to_categorical(train_df['ATYPE'].to_numpy())
```

```python
x_test = test_df[columns].to_numpy()
y_test = to_categorical(test_df['ATYPE'].to_numpy())
```

### Establecemos las estructuras de almacenamiento e iteración de nuestro modelo

```python
batch_size = [128]
epochs = [25]
perceptrons = [32, 64, 128, 256]
dropout = [0.05]
xfeatures = [64, 256]
kernel_size = [2]
pool_size = 2
cdense = 128
```

```python
model_result = {'conf': None, 'xaccuracy': None, 'yaccuracy': None, 'xloss' : None, 'yloss' : None, 
                'bepoch' : None, 'baccuracy': None, 'bloss': None}
models_analysis = list()
```

### Importamos las clases Keras necesarias para la creacion de nuestra MLP básica y componemos el modelo.

```python
input_dimension = len(train_df.columns)-1
output_dimension = len(train_df['ATYPE'].unique())
```

```python
c_train = x_train.reshape([-1, 6, 5, 1])
c_test = x_test.reshape([-1, 6, 5, 1])
cinput_dimension = (c_train.shape[1], c_train.shape[2], 1)
```

```python
for hl11 in perceptrons:
    for hl12 in perceptrons: 
        for hl1 in xfeatures:
                for bs in batch_size:
                    for ks in kernel_size:
                        for ep in epochs:
                            for dp in dropout:
                                input_1 = Input(shape=(input_dimension,))
                                x = input_1
                                x = Dense(hl11)(x)
                                x = Activation('relu')(x)
                                x = Dropout(dp)(x)
                                x = Dense(hl12)(x)
                                x = Activation('relu')(x)
                                x = Dropout(dp) (x)
                                
                                input_2 = Input(shape=cinput_dimension)
                                y = input_2
                                y = Conv2D(filters=hl1, kernel_size=ks, activation='relu', 
                                     data_format='channels_last',padding='same')(y)
                                y = MaxPooling2D(pool_size)(y)
                                y = Flatten()(y)
                                y = Dense(cdense)(y)
                                y = Activation('relu')(y)
                                y = concatenate([x,y])
                                
                                output = Dense(output_dimension, activation = 'softmax' ) (y)
                                model = Model([input_1, input_2], output)

                                mr = model_result.copy()
                                mr['conf'] = [[hl11, hl12],[hl1,ks, pool_size], bs, ep, dp]            

                                # loss function for one-hot vector
                                # use of adam optimizer
                                # accuracy is a good metric for classification tasks
                                model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

                                # train the network
                                model_history = model.fit(x=[x_train, c_train], y=y_train, batch_size=bs, epochs=ep)
                                index_acc = argmax(model_history.history['acc'])
                                mr['bloss'] = model_history.history['loss'][-1]
                                mr['baccuracy'] = model_history.history['acc'][-1]
                                mr['bepoch']= index_acc + 1
                                mr['xloss'] = model_history.history['loss'][index_acc]
                                mr['xaccuracy'] = model_history.history['acc'][index_acc]

                                # Evaluate the network                                                             
                                mr['yloss'], mr['yaccuracy'] = model.evaluate([x_test, c_test],y_test, batch_size=bs)
                                models_analysis.append(mr)   
```

### MLP best network

```python
best_model = None

for itera in models_analysis:
    if best_model is None:
        best_model = itera
    elif itera['baccuracy'] > best_model['baccuracy']:
        best_model = itera

print (best_model)
```

### Testing code MLP

```python
input_1 = Input(shape=(input_dimension,))
x = input_1
x = Dense(best_model['conf'][0][0])(x)
x = Activation('relu')(x)
x = Dropout(best_model['conf'][4])(x)
x = Dense(best_model['conf'][0][1])(x)
x = Activation('relu')(x)
x = Dropout(best_model['conf'][4]) (x)
                                
input_2 = Input(shape=cinput_dimension)
y = input_2
y = Conv2D(filters=best_model['conf'][1][0], kernel_size=best_model['conf'][1][1], 
           activation='relu', data_format='channels_last',padding='same')(y)
y = MaxPooling2D(best_model['conf'][1][2])(y)
y = Flatten()(y)
y = Dense(cdense)(y)
y = Activation('relu')(y)
y = concatenate([x,y])
                                
y = concatenate([x,y])
                                
output = Dense(output_dimension, activation = 'softmax' ) (y)
model = Model([input_1, input_2], output)

# loss function for one-hot vector
# use of adam optimizer
# accuracy is a good metric for classification tasks
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# train the network
model_history = model.fit(x=[x_train, c_train], y=y_train, batch_size=best_model['conf'][2], epochs=best_model['bepoch'])

a,b = model.evaluate([x_test,c_test],y_test, batch_size=best_model['conf'][2])
```
